#### Convertendo Imagens para arrays Numpy

```
#carregando imagem como um array NumPy

matriz_img = image.imread("/content/Teste LSTF 16 6.jpg.jpg")
```

###Cortar Imagens (Crop)

```
#exibindo o corte a partir das coordenadas especificadas - left, upper, right, and lower

area = (315,290,472,516)
cort_img = imagem.crop(area)
cort_img
```

###Convertendo imagens

```
#convertendo a imagem em escala de cinza

imagem_cinza = cort_img.convert(mode="L")
imagem_cinza
```

```
print(imagem_cinza.size)
```

```
imagem_cinza.save("img_cinza.jpg")
```
